```python
import re
import yaml
import os
```

```python
yaml_file = open("SkyTeam-Exchange.yaml", "r")
```

```python
newFile = None
for line in yaml_file:
    isDate = re.search(r'\d{4}-\d{2}-\d{2}', line)
    if isDate:
        if newFile:
            newFile.close()
        newFile = open('resultDataParse/' + isDate[0] + '.yaml', mode = 'w')
    else:
        newFile.write(line)
```

```python
writeFile = open("SkyTeam-Exchange.csv", "w")
#Headers
writeFile.write("code;date;departure;status;arrival;cardNumber;class;fare\n")
count = 0
for fileName in os.listdir("resultDataParse"):
    count = count + 1
    print(str(count) + " " + fileName)
    openFile = open("resultDataParse/" + fileName, "r")
    yaml_yaml = yaml.load(openFile)
    for main_key, main_value in yaml_yaml.items():
        for inner_key, inner_value in main_value['FF'].items():
            departure = "null"
            status = "null"
            arrival = "null"
            classValue = "null"
            fare = "null"
            code = main_key
            date = fileName.split(".")[0]
            if 'FROM' in main_value:
                departure = main_value['FROM']
            if 'STATUS' in main_value:
                status = main_value['STATUS']
            if 'TO' in main_value:
                arrival = main_value['TO']
            cardNumber = inner_key
            if 'CLASS' in inner_value:
                classValue = inner_value['CLASS']
            if 'FARE' in inner_value:
                fare = inner_value['FARE']
            writeFile.write(code + ";" + date + ";" + departure + ";" + status + ";" + arrival + ";" + cardNumber + ";" + classValue + ";" + fare + "\n")
    openFile.close()
writeFile.close()
```
